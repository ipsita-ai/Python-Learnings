```python
def multiply(a,b):
    result = 0
    for i in range(0,a):
        result += b
    return result
```

```python
print(multiply(10022,102))
```

```python
# For recursion you identify base case
# 3*4 = 4 + 2*4 = 4 + 4 + 1*4
# a*b ==> a = 1 ==>a*b = b
```

```python
def mul(a,b):
    if a == 1:
        return b
    else:
        return mul(a-1,b) + b
```

```python
mul(102,102)
```

```python
# Find the factorial of a given number 
#5 = 5*4*3*2*1

def fact(num):
    if num == 1:
        return 1
    else:
        return num*fact(num-1)
```

```python
fact(5)
```

```python
# Palindrome
# if the string is single charecter is palindrome --> base condition
# abba = abba
# sas = sas
# madam = madam
```

```python
def palin(string):
    if len(string) <= 1:
        print('Palindrome')
    else:
        if string[0] == string[len(string)-1]:
            palin(string[1:-1])
        else:
            print('Not a Palindrome')
```

```python
palin('madam')
```

```python
palin('abba')
```

```python
# The rabit problem/Fibonacci problem
# 1,1,2,3,5,8,13
# Time complexity is 2^n
```

```python
def fib(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python
fib(37)
```

```python
# memoization --> Space Time Trade Off
```

```python
def memo(num):
    d = {1:1,2:1}
    if num in d:
        return d[num]
    else:
        d[num] = memo(num-1) + memo(num-2)
        return d[num]
```

```python
memo(40)
```

#### power set



```python
def power_set(input_set):
    if len(input_set) == 0:
        return [[]]
    else:
        first_elem = input_set[0]
        rest_of_set = input_set[1:]

        # Recursive call to get the power set of the rest of the set
        power_set_rest = power_set(rest_of_set)

        # Add the first element to each subset in the power set of the rest
        with_first_elem = [([first_elem] + subset) for subset in power_set_rest]

        # Combine the subsets with and without the first element
        result = with_first_elem + power_set_rest

        return result

```

```python
power_set([1,2,3])
```
