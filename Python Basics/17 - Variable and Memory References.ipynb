```python
# How values are stored in RAM
```

```python
a = 4
print(hex(id(4)))
print(hex(id(a)))
```

```python
# Aliasing
a = 5
b = a
print(id(a) == id(b))
del a 
print(b)
a = 5
b = a
b = 6
print(a,b)
print(id(a) == id(b))
```

```python
# Reference counting
import sys 
m = 'assdcsdfsdf'
b = m
c = b
print(id(m)==id(b)==id(c))
sys.getrefcount(m)
```

```python
# Garbage Collection ---> python runs a garbage collector which cleans off unused data in RAM
```

```python
a = [1,2,3,4]
b = a[:]
b.append(5)
print(id(a) == id(b))
print(a,b)
```

```python
# Shallow copy
a = [1,2,3,4,[5,6]]
b = a[:]
print(b,a)
b[-1][0] = 7
print(b,a)
```

```python
# Shallow copy
a = [1,2,3,4,[5,6]]
b = a
print(b,a)
b[-1][0] = 7
print(b,a)
```

```python
# Shallow copy
a = [1,2,3,4,[5,6]]
b = a.copy()
print(b,a)
b[-1][0] = 7
print(b,a)
```

```python
# Deep copy
import copy
a = [1,2,3,4,[5,6]]
b = copy.deepcopy(a)
b[-1][0] = 7
print(a,b)
```

```python
a = 'Birla Institute Of Technology'
b = 'Birla Institute Of Technology'
id(a) == id(b)
```

```python
# Valid identifier
a = 'Birla_Institute_Of_Technology'
b = 'Birla_Institute_Of_Technology'
id(a) == id(b)
```

```python
# Mutability
l = [1,2,3,4]
id(l[0])==id(1)
l[3] = 1
id(l[0]) == id(l[3])
```

```python
a = [1,2]
b = [3,4]
c = (a,b)
print(c,id(c))
a.append(5)
b.append(6)
print(c,id(c))
c = c + (7,8,9)
print(c,id(c))
```
