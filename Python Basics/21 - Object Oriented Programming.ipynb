```python
class Atm:
    # Variables are only declared under init function which is a special function.
    # INIT is a constructor is a special method which is executed as soon as you create a object of the class
    # Constructor : Special method , magic method, dunder method
    # Constructors are not called by object , they get automatically triggered
    # We usually have configuration based task whose control is not given to user inside a constructor 
    def __init__(self):
        self.pin =''
        self.balance = 0
        print(id(self))
        self.menu()
        
    def menu(self):
        user_input = input('''
        Hello , How would you like to proceed ?
            1- Enter 1 to create pin
            2- Enter 2 to deposit
            3- Enter 3 to withdraw
            4- Enter 4 to check balance
            5- Enter 5 to exit
            
        ''')
        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.deposit()
        elif user_input == '3':
            self.withdraw()
        elif user_input == '4':
            self.check_balance()
        else:
            print('Thank You')
            
    def create_pin(self):
        self.pin = input('Enter your pin to set it')
        print('Pin Set Successfully')
        self.menu()
    def deposit(self):
        temp = input('Enter your pin')
        if self.pin == temp:
            amount = int(input('Enter the amount to be deposited : '))
            self.balance = self.balance + amount
            print('Deposit successful')
        else:
            print('Invalid Pin')
        self.menu()
    def withdraw(self):
        temp = input('Enter your pin')
        if self.pin == temp:
            amount = int(input('Enter the amount to be withdrawn :'))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print('Withdraw successful')
            else:
                print('Insufficient Balance')
        else:
            print('Invalid Pin')
        self.menu()
    def check_balance(self):
        temp = input('Enter your pin')
        if self.pin == temp:
            print(self.balance)
        else:
            print('Invalid Pin')
        self.menu()
```

```python
class Fraction:
    def __init__(self,n,d):
        self.num = n
        self.den = d
    def __str__(self):
        return '{}/{}'.format(self.num,self.den)
    def __add__(self,other):
        numerator = self.num*other.num + self.den*other.num
        denomerator = self.den*other.den
        return '{}/{}'.format(numerator,denomerator)
    def __sub__(self,other):
        numerator = self.num*other.num - self.den*other.num
        denomerator = self.den*other.den
        return '{}/{}'.format(numerator,denomerator)
    def __mul__(self,other):
        numerator = self.num*other.num
        denomerator = self.den*other.den
        return '{}/{}'.format(numerator,denomerator)
    def __truediv__(self,other):
        numerator = self.num*other.den
        denomerator = self.den*other.num
        return '{}/{}'.format(numerator,denomerator)    
```

```python
## Static Object 
## In Python programming, a static variable is said to be a class variable that is common to all class members.
## A static variable is declared within that class but outside the methods within the given class.
## Non-static variables are those variables that are associated with the objects or methods declared within the class.

class Atm:
    # static/class variable
    __counter = 1
    def __init__(self):
        # instance variable
        self.__pin =''
        self.__balance = 0
        self.__menu()
        self.sno = Atm.__counter
        Atm.__counter += 1
    ## Static Method
    @staticmethod
    def get_counter():
        return Atm.__counter
    @staticmethod
    def set_counter(value):
        if type(value) == int:
            Atm.__counter = value
        else:
            print('Not allowed')
    def get_pin(self):
        return self.__pin
    def set_pin(self,new_pin):
        self.__pin = new_pin
        print('pin changed')
    def __menu(self):
        user_input = input('''
        Hello , How would you like to proceed ?
            1- Enter 1 to create pin
            2- Enter 2 to deposit
            3- Enter 3 to withdraw
            4- Enter 4 to check balance
            5- Enter 5 to exit
            
        ''')
        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.deposit()
        elif user_input == '3':
            self.withdraw()
        elif user_input == '4':
            self.check_balance()
        else:
            print('Thank You')
            
    def create_pin(self):
        self.__pin = input('Enter your pin to set it')
        print('Pin Set Successfully')
        self.__menu()
    def deposit(self):
        temp = input('Enter your pin')
        if self.__pin == temp:
            amount = int(input('Enter the amount to be deposited : '))
            self.__balance = self.__balance + amount
            print('Deposit successful')
        else:
            print('Invalid Pin')
        self.__menu()
    def withdraw(self):
        temp = input('Enter your pin')
        if self.__pin == temp:
            amount = int(input('Enter the amount to be withdrawn :'))
            if amount <= self.__balance:
                self.__balance = self.__balance - amount
                print('Withdraw successful')
            else:
                print('Insufficient Balance')
        else:
            print('Invalid Pin')
        self.__menu()
    def check_balance(self):
        temp = input('Enter your pin')
        if self.__pin == temp:
            print(self.__balance)
        else:
            print('Invalid Pin')
        self.__menu()
```

```python
# Aggregation
class Customer:
    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender
        self.address = address
    def edit_profile(self,new_name,new_city,new_pin,new_country):
        self.name = new_name
        self.address.change_address(new_city,new_pin,new_country)
class Address:
    def __init__(self,city,pincode,country):
        self.city = city
        self.pincode = pincode
        self.country = country
    def change_address(self,new_city,new_pin,new_country):
        self.city = new_city
        self.pincode = new_pin
        self.country = new_country
add = Address('Paris',78500,'France')
cust = Customer('Ipsita','Female',add)
```

```python
# Child class inherits the constructor of parent class
class Phone:
    def __init__(self,price,brand,camera):
        print('inside phone constructor')
        self.price = price
        self.brand = brand
        self.camera = camera
class SmartPhone(Phone):
    pass

s = SmartPhone(20000,'Apple',13)
print(s.brand)
```
