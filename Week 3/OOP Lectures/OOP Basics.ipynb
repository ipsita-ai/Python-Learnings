```python
# To create class name we use pascal case
class Atm:
    # Constructur is a special function inside the class-->superpower
    # Code inside the constructor is executed with out explicit calling.The moment object of class is created constructor is automatically executed
    # Variables inside the class should be defined inside constructor
    # To use the variable we should use self
    def __init__(self): # non-parameterized constructor
        print(id(self))
        self.pin = ''
        self.balance = ''
        #self.menu()
    def menu(self):
        user_input = input('''
        Hi , How can I help you ?
        Choose from the options below :
        1.Set Pin
        2.Change Pin
        3.Withdraw Amount
        4.Check Balance
        5.Exit
        ''')
        if user_input == '1':
            #set pin
            self.set_pin()
        elif user_input == '2':
            #change pin
            self.change_pin()
        elif user_input == '3':
            #withdraw amount
            self.withdraw()
        elif user_input == '4':
            #Check Blance
            self.check_balance()
        else:
            exit()

    def set_pin(self):
        user_pin = input('Enter your Pin')
        self.pin = user_pin
        user_balance = int(input('Enter your Balance'))
        self.balance = user_balance
        print('Pin and Balance set successfully')
        self.menu()
    def change_pin(self):
        old_pin = input('Enter existing pin')
        if old_pin == self.pin:
            #change pin allowed
            new_pin = input('Enter new pin')
            self.pin = new_pin
            print('Pin changed successfully')
        else:
            print('Incorrect Pin')
        self.menu()
    def check_balance(self):
        pin_check = input('Enter pin')
        if pin_check == self.pin:
            print('Your balance is ,',self.balance)
        else:
            print('Incorrect pin entered')
            self.check_balance()
        self.menu()
    def withdraw(self):
        pin_check = input('Enter pin')
        if pin_check == self.pin:
            withdraw_balance = int(input('Enter withdrawal amount'))
            if withdraw_balance <= self.balance:
                self.balance = self.balance - withdraw_balance
                print('Withdrawal successful')
                print('Your new balance is ',self.balance)
            else:
                print('Insufficient amount')
        else:
            print('Incorrect pin')
            self.withdraw()
        self.menu()
```

```python
# Fraction Data Type
class Fraction:
    # Parameterized constructor
    def __init__(self,x,y):
        self.num = x
        self.den = y 
    def __str__(self):
        # when object is put inside print --> print tries to find __str__ to understand how to show the object
        return '{}/{}'.format(self.num,self.den)
    def __add__(self,other):
        #triggered when + operator is used
        return '{}/{}'.format((self.num*other.den + self.den*other.num),self.den*other.den)
    def __sub__(self,other):
        #triggered when - operator is used
        return '{}/{}'.format((self.num*other.den - self.den*other.num),self.den*other.den)
    def __mul__(self,other):
        #triggered when * operator is used
        return '{}/{}'.format(self.num*other.num,self.den*other.den)
    def __truediv__(self,other):
        #triggered when * operator is used
        return '{}/{}'.format(self.num*other.den,self.den*other.num)
    def convert_to_dec(self):
        return self.num/self.den
```
