###### Operators in Python

- Arithmetic Operator
- Relational Operator
- Logical Operator
- Bitwise Operator
- Assignment Operator
- Membership Operator

```python
# Assignment Operator
# Summation
print(5 + 6)
# Subtraction
print(5 - 6)
# Multiplication
print(5 * 6)
# Division
print(5 / 6)
# Integer Division
print(5 // 2)
# Remainder (Modulus operator)
print(5 % 6)
# Power of 
print(5 ** 2)
```

```python
# Relational Operators
print(4 > 5)
print(4 < 5)
print(4 >= 4)
print(4 <= 4)
print(4 != 4)
```

```python
# Logical Operators
# and - 1 and 1 is 1 , 1 and 0 is 0 , 0 and 1 is 0 , 0 and 0 is 0
print(1 and 0)
# or - 1 and 1 is 1 , 1 and 0 is 1 , 0 and 1 is 1 , 0 and 0 is 0
print(1 or 0)
# not
print(not 1)
```

```python
# Bitwise operators
# Bitwise and
# 2 - 10 , 3 - 11 == 2&3 - 10 
print(2 & 3)
# Bitwise or 
# 2 - 10 , 3 - 11 == 2|3 - 11 
print(2|3)
# Bitwise xor - 1,1 and 0,0 is 0 and 0,1 and 1,0 is 1
# 2 - 10 , 3 - 11 == 2|3 - 01 
print(2^3)
# Bitwise not 011 to 100
print(~3)
# Bitwise Left Operator
print(4 >> 1)
# Bitwise Right Operator
print(5 << 1)
```

```python
# Assignment Operators

a = 3
a += 4
print(a)
a //= 2
print(a)
```

```python
# Membership Operators
# in / not in 
print('d' in 'delhi')
print('m' not in 'Mumbai')
print(1 in [1,2,3,4])
```

```python
# Program - Find the sum of a 3 digit number entered by the user 

f_num = int(input('Enter any three digit number : '))

sum_num = f_num % 10 + (f_num //10)%10 + f_num //100

print('The sum of digits in {} is {}'.format(f_num,sum_num))
```

##### If else in Python

```python
# If else allows for branching / decision making 
# Login Program
# Email : ipsita.panda@edhec.com
# Password : 1234

email = input('Enter email : ')
password = input('Enter password')

if email == 'ipsita.panda@edhec.com' and password == '1234':
        print('Welcome to our website')
elif email == 'ipsita.panda@edhec.com' and password != '1234':
    print('Incorrect password.Enter your password again')
    password = input('Enter password')
    # Nested If else
    if password == '1234':
        print('Welcome to our website')
    else:
        print('Incorrect password.Reset your password')
else:
    print('Incorrect username or password')
    
```

```python
# If else examples
# Find min of 3 numbers
num_1 = int(input('Enter first number : '))
num_2 = int(input('Enter first number : '))
num_3 = int(input('Enter first number : '))

if num_1 < num_2 and num_1 < num_3:
    print('The mininum is',num_1)
elif num_2 <num_1 and num_2 < num_3:
    print('The mininum is',num_2)
else:
    print('The mininum is',num_3)
```

```python
# Menu Driven Program 
# Example Calculator
fnum = int(input('Enter the first number : '))
snum = int(input('Enter the second number : '))
op = input('Enter the operator : + - * / : ' )
if op == '+':
    print(fnum + snum)
elif op == '-':
    print(fnum - snum)
elif op == '*':
    print(fnum*snum)
elif op == '/':
    print(fnum / snum)
```

###### Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
import math
current_population = 10000
for i in range(10,0,-1):
    population = math.floor(current_population/1.1)
    current_population = population
    print('Year',i,',the population is',population)
```

```python
## Sequence sum 
#1/1! + 2/2! + 3/3! + 4/4!
result = 0
fact = 1
n = int(input('Enter the value of n for the sequence'))
for i in range(1,n+1):
    fact *= i
    result += i/fact
print(result)
```

```python
# Pattern 2
1
121
12321
1234321
123454321

n = int(input('Enter the number of rows : '))
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end= '')
    for k in range(i-1,0,-1):
        print(k,end='')
    print()
```

```python
for i in range(10):
    if i == 5:
        break
    print(i)
    
```

```python
# Find prime numbers in a user given range
lwr_rng = int(input('Enter lower range : '))
upr_rng = int(input('Enter upper range : '))
for i in range(lwr_rng,upr_rng + 1):
    for j in range(2,i):
        if i%j == 0:
            break
    else:
        print('{} is a prime number'.format(i))
```
