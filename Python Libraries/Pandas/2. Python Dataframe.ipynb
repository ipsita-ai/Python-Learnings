```python
# write a function that can return the track record of 2 teams against each other
def compare_teams(team1,team2):
    mask_team1 = (ipl['Team1'] == team1)
    mask_team2 = (ipl['Team2'] == team2)
    data = ipl[mask_team1 & mask_team2]
    return data['WinningTeam'].value_counts()
    
   
        
    
```

```python
compare_teams('Chennai Super Kings','Mumbai Indians')
```
